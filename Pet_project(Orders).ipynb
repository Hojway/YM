```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Динамика продаж по месяцам

```python
orders = pd.read_excel('Downloads/united_orders_21597717_01-01-2023_29-05-2023.xlsx')
orders
```

```python
orders.fillna(0, inplace=True)
orders
```

```python
orders.columns
```

```python
orders['Сумма покупки'] = orders['Количество'] * orders['Ваша цена (за шт.)']
orders[['Сумма покупки', 'Количество', 'Ваша цена (за шт.)']].head(20)
```

```python
orders['Дата оформления'] = pd.to_datetime(orders['Дата оформления'], format='%d.%m.%Y')
```

```python
orders['Статус товара'].unique()
```

```python
month_orders = orders[(orders['Дата оформления'] >= '01.01.2023') & (orders['Дата оформления'] < '06.01.2023')
                 & (orders['Статус товара'].isin(['Доставлен', 'В обработке', 'Доставлен на пункт выдачи', 'Передан в доставку', 'Оформлен, но не оплачен']))]
month_orders['Сумма покупки'].sum()
```

```python
results = month_orders.groupby(month_orders['Дата оформления'].dt.month)['Сумма покупки'].sum()
results
```

```python
months = range(1,6)

barplot = plt.bar(months, results)
plt.title('Monthly revenue')
plt.xlabel('Months')
plt.ylabel('Rubles')
plt.bar_label(barplot, labels=results)
plt.show()
```

#### Анализ конверсии продаж по SKU

```python
displays = pd.read_excel('Downloads/displays_and_sells_goods_21597717_01.05.23-29.05.23.xlsx')
displays
```

```python
displays.fillna(0, inplace=True)
```

```python
sku_dis = displays.groupby('Ваш SKU', as_index = False).sum()
sku_dis
```

```python
may_rev = (month_orders[month_orders['Дата оформления'].dt.month==5]).groupby('Ваш SKU', as_index=False).sum()

may_ord = may_rev[['Ваш SKU', 'Количество']]
may_ord
```

```python
sku_dis_ord = pd.merge(sku_dis,
 may_ord,
 on ='Ваш SKU',
 how ='left')
sku_dis_ord
```

```python
sku_dis_ord['Конверсия продаж'] = sku_dis_ord['Показы']/sku_dis_ord['Количество']
sku_dis_ord.sort_values(by=['Конверсия продаж'])[['Ваш SKU','Конверсия продаж']]
```

#### Этот расчет показывается конверсия продаж каждой SKU, чем меньше значение тем выше конверсия
